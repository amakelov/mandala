```python
from mandala.all import *
storage = Storage(in_memory=True)
```

```python
@op(storage)
def add(x, y) -> int:
    print('Hello world!')
    # ...some long computation...
    return x + y 
```

```python
add(23, 42)
```

```python
with run(storage):
    result = add(23, 42)
result
```

```python
with run(storage):
    result = add(23, 42)
result
```

```python
with run(storage):
    for n in range(20, 25):
        result = add(n, 42)
```

```python
@op(storage)
def inc(x) -> int:
    print('Hello from inc!')
    return x + 1 
```

```python
with run(storage):
    for n in range(3):
        n_inc = inc(n)
        result = add(n, n_inc)
```
